```python
import psycopg2
ma_base_donnees = "pollution_BD"
utilisateur = "postgres"
mot_passe = os.environ.get('pg_psw')

conn = postgreSQL_functions.ouvrir_connection(ma_base_donnees, utilisateur, mot_passe)
```

```python
# Créer les tables station, polluant No1, polluant PM10
```

```python
sql_creation_table_station = """
    CREATE TABLE IF NOT EXISTS station (
    code_station INTEGER primary key,
    nom_station text,
    nom_commun text,
    nom_departement text,
    typologie text,
    x_reglementaire float,
    y_reglementaire float);
"""

postgreSQL_functions.creer_table(conn, sql_creation_table_station)
```

```python
sql_creation_table_polluant= """
    CREATE TABLE IF NOT EXISTS polluant (
    id_polluant SERIAL primary key,
    polluant_nom text);
"""
postgreSQL_functions.creer_table(conn,sql_creation_table_polluant)
```

```python
sql_creation_table_mesure_journallier = """
    CREATE TABLE IF NOT EXISTS mesure_journallier (
    id SERIAL,
    Niveau FLOAT,
    Date DATE,
    etat text,
    id_polluant INTEGER,
    code_station INTEGER,
    FOREIGN KEY (id_polluant) REFERENCES polluant(id_polluant),
    FOREIGN KEY (code_station) REFERENCES station(code_station));
"""
postgreSQL_functions.creer_table(conn,sql_creation_table_mesure_journallier)
```

```python
# pour inserer les données dans la table mesure journalliere, on va importer deux fichier csv en forme de panda data frame,
# les nettoyer et combiner a fin de les inserer dans la table pollution:
mesures_NO2=pandas.read_csv('data/mesures_NO2.csv',sep=";")
```

```python
mesures_NO2.head(n=5)
```

```python
mesures_NO2_nettoyer=mesures_NO2.iloc[:,[3,4,5,8,9]]
mesures_NO2_nettoyer['polluant']=mesures_NO2_nettoyer['polluant'].str.replace('NO2','8')
mesures_NO2_nettoyer['polluant']=pandas.to_numeric(mesures_NO2_nettoyer['polluant'])
```

```python
mesures_PM10=pandas.read_csv('data/mesures_PM10.csv',sep=";")
mesures_PM10_nettoyer=mesures_PM10.iloc[:,[3,4,5,8,9]]
mesures_PM10_nettoyer['polluant']=mesures_PM10_nettoyer['polluant'].str.replace('PM10','5')
mesures_PM10_nettoyer['polluant']=pandas.to_numeric(mesures_PM10_nettoyer['polluant'])
```

```python
# vue que on va servire de la même requete sql qui retourne des informations sur des stations pour une date données 
# et une polluant defini, on a créer la fonction ci-après qui prend en entrée id de polluant ainsi une condition sur la date 
# et/ou sur le niveau de pollution. il retourne des informations sur les stations
# attention quand vous entrer une date dans cette fonction, BETWEEN ne marche pas, il faut difinir la date comme l'exemple
# "date>='yyyy-mm-jj' AND date<='yyyy-mm-jj'"

def requete_fonction(id_polluant,condition_date_niveau):
    requete=f"SELECT station.code_station,nom_station,nom_commun,nom_departement,typologie,niveau,date,etat FROM mesure_journallier LEFT JOIN station ON mesure_journallier.code_station=station.code_station WHERE id_polluant={id_polluant} AND {condition_date_niveau};"
    return(requete)
    
```

```python
# Etudier l’influence des confinements sur la qualité de l’air dans la région en fonction des départements et de deux types de polluants
# 
```

```python
# list nom station
list_nom_station=station_donnes_csv.drop_duplicates(["code_station"])
list_nom_station=list_nom_station["nom_station"]
#list_nom_station
```

```python
# plot_fonction prend en entre pandas data frame, liste de noms de stations est retourn une grahique date/niveau polution
# pour chaque station donnée dans le variable liste_nom_station:
def plot_fonction(df_qualite_air,liste_nom_station,title_plot):
    plt.figure(figsize=(20,10))
    pl=plt.plot()
    for i in liste_nom_station:
        z=df_qualite_air.loc[df_qualite_air.nom_station==i]
        x = z['date']
        y = z['niveau']
        pl=pl+plt.plot(x,y,label=i)
        plt.xlabel("Date",fontsize=14)
        plt.ylabel("niveau PM10",fontsize=14)
        plt.xticks(rotation = '90',fontsize=12)
        plt.title(title_plot,fontsize=16)
        plt.grid(True)
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
#qualite d'air en 2019, niveau PM10:

qualite_air_PM10_2019=requete_fonction(5,"date>='2019-01-01' AND date<='2019-12-30'")

qualite_airPM10_2019 = pandas.read_sql_query(qualite_air_PM10_2019, conn)

# Visualisation des données: 
plot_fonction(qualite_airPM10_2019,
              list_nom_station,
              "Qualite de l'air au niveau de PM10 en 2019")
plt.ylim(0,70)
```

```python
#qualite d'air en 2020 en visualisant les deux periode de confinement:

qualite_airPM10_2020=requete_fonction(5,"date>='2020-01-01' AND date<='2020-12-30'")
qualite_airPM10_2020 = pandas.read_sql_query(qualite_air_PM10_2020, conn)
# Visualisation des données: 
plot_fonction(qualite_airPM10_2020,
              list_nom_station,
              "Qualite de l'air au niveau de PM10 en 2020")
plt.ylim(0,70)
plt.axvline(pandas.Timestamp('2020-03-17'),color='r',linewidth=2)
plt.axvline(pandas.Timestamp('2020-05-10'),color='r',linewidth=2)
plt.text(pandas.Timestamp('2020-03-17'),68,'confinement', color='red',fontsize=14)
plt.axvline(pandas.Timestamp('2020-10-30'),color='r',linewidth=2)
plt.axvline(pandas.Timestamp('2020-12-15'),color='r',linewidth=2)
plt.text(pandas.Timestamp('2020-10-30'),68,'confinement', color='red',fontsize=14)
plt.text(pandas.Timestamp('2019-10-15'),50,'Seuil recommande', color='red',fontsize=14)
```
